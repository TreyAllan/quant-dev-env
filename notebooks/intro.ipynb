## Step 1: Generate Synthetic Time-Series Data

In this step, we'll generate a synthetic time-series dataset using **pandas** and **numpy**. This data represents random values over a set period (100 days) with the date range starting from `2023-01-01`.

We'll simulate this as though it's financial data or any other time-dependent data that might need analysis.

```python
import pandas as pd
import numpy as np

# Generate some fake time-series data
dates = pd.date_range('20230101', periods=100)  # 100 days starting from Jan 1, 2023
data = np.random.randn(100, 1)  # Random numbers to simulate some data

df = pd.DataFrame(data, index=dates, columns=['Value'])  # Store in a pandas DataFrame

# Show the first few rows of the dataframe
df.head()
```

- The `pandas.date_range` function creates the **dates** for the time-series.
- The `numpy.random.randn` generates 100 random values, simulating daily returns or values.
- These values are then stored in a **DataFrame** with `dates` as the index and a single column labeled `'Value'`.

At this point, we have a simple DataFrame with a random time series.

---

### Step 2: Visualize the Time-Series Data
Now let's visualize the generated time-series data to better understand its trend. In this step, we'll use **matplotlib** to plot the synthetic time-series data. This will allow us to visually inspect the generated values and see if any patterns emerge.

We'll use the **line plot** to represent the data, where the x-axis will be the dates and the y-axis will show the random values generated.

```python
import matplotlib.pyplot as plt

# Plot the data
df.plot()

# Customize the plot
plt.title('Random Time-Series Data')  # Title
plt.xlabel('Date')  # X-axis label
plt.ylabel('Value')  # Y-axis label
plt.grid(True)  # Show grid
plt.show()
```

- The `df.plot()` function automatically generates a line plot.
- We added a title and axis labels for better clarity.
- The `plt.grid(True)` command adds a grid to the plot, making it easier to read.

The plot should show random fluctuations over the 100-day period, simulating how financial data might look on a basic level.
